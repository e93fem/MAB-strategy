```python
sessions_first = [x for x in log_file_first if x.startswith("Total result:")]
sessions_last = [x for x in log_file_last if x.startswith("Total result:")]
sessions_evenly = [x for x in log_file_evenly if x.startswith("Total result:")]
sessions_ageing_first = [x for x in log_file_ageing_first if x.startswith("Total result:")]
sessions_ageing_last = [x for x in log_file_ageing_last if x.startswith("Total result:")]
```

```python
sessions_first[0:4]
```

```python
sessions_val_first = pd.Series([float(x.split(" ")[7]) for x in sessions_first])
sessions_val_last = pd.Series([float(x.split(" ")[7]) for x in sessions_last])
sessions_val_evenly = pd.Series([float(x.split(" ")[7]) for x in sessions_evenly])
sessions_val_ageing_first = pd.Series([float(x.split(" ")[7]) for x in sessions_ageing_first])
sessions_val_ageing_last = pd.Series([float(x.split(" ")[7]) for x in sessions_ageing_last])
```

```python
min_size = min(len(sessions_val_first), len(sessions_val_last), len(sessions_val_evenly), len(sessions_val_ageing_first), len(sessions_val_ageing_last))
```

```python
plt.figure(1)
ax = plt.subplot(1, 1, 1)
plt.plot(sessions_val_first[:min_size-1], label='First')
plt.plot(sessions_val_last[:min_size-1], label='Last')
plt.plot(sessions_val_evenly[:min_size-1], label='Evenly')
plt.plot(sessions_val_ageing_first[:min_size-1], label='Ageing First')
plt.plot(sessions_val_ageing_last[:min_size-1], label='Ageing Last')
plt.legend(loc='upper left')
plt.show()
```
